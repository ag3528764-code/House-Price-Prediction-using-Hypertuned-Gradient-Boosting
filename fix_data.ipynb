```python
import pandas as pd
from sklearn.model_selection import train_test_split

def generate_clean_split_datasets():
    # ── STEP 1: LOAD RE-WRITTEN RAW DATASETS ──────────────────
    try:
        # Hamesha original full backup dataset load hoga jisme target variable available hai
        df = pd.read_csv('train.csv')
        print(f"🚀 [DATA SETUP] Raw spreadsheet loaded successfully. Total Records: {len(df)} rows.")
    except FileNotFoundError:
        print("❌ Error: 'train.csv' file not found in the current root directory.")
        print("💡 Hint: Make sure your original un-split Kaggle dataset is named 'train.csv'.")
        return

    # ── STEP 2: VERIFY VALID SPLIT CONDITIONS ──────────────────
    # Check if the dataset is already sliced or needs a proper split
    if len(df) < 500:
        print("⚠️ Warning: Your 'train.csv' is already containing a small sliced chunk.")
        print("💡 Action: Please replace 'train.csv' with your original un-split full dataset file and run again.")
        return

    # ── STEP 3: EXECUTE RANDOM SHUFFLE SPLIT ────────────────────
    print("🎲 Shuffling and splitting datasets randomly to eliminate ordering bias...")
    
    # 80% matrix for production training loop, 20% dedicated for metric evaluation
    train_split, test_split = train_test_split(df, test_size=0.20, random_state=42, shuffle=True)

    # ── STEP 4: OVERWRITE DATASETS ON DISK ─────────────────────
    print("💾 Exporting balanced vectors back into file sheets...")
    
    # Writing matrices back to local disk safely
    train_split.to_csv('train.csv', index=False)
    test_split.to_csv('test.csv', index=False)

    print("\n🎯 [SUCCESSFUL PRODUCTION RUN] 🎯")
    print(f"• Generated new 'train.csv' matrix shape: {train_split.shape[0]} rows (For Training)")
    print(f"• Generated new 'test.csv'  matrix shape: {test_split.shape[0]} rows (With 'SalePrice' target for Testing)")
    print("-" * 75)
    print("👉 Next Action Steps:")
    print("1. Execute 'python train_and_plot.py' to fit pipelines on the new random rows.")
    print("2. Execute 'python predict_test.py' to view your high-accuracy true R² and MAE evaluation scores!\n")

if __name__ == "__main__":
    generate_clean_split_datasets()
```
